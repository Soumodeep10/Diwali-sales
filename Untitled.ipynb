```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status' , 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
ax = sns.countplot(x='Gender', hue='Gender', data=df, palette='Set2', legend=False)


for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
print(sales_gen)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette='Set2', legend=False)
plt.show()
```

```python
ax = sns.countplot(x='Age Group', hue='Gender', data=df)


for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sns.countplot(data=df,x='Age Group', hue='Gender', )
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group')
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State', y='Orders', data=sales_state, hue='State')
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(x='State', y='Amount', data=sales_state, hue='State')
plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=sales_state, hue='Occupation')
plt.show()
```

```python
ax = sns.countplot(x='Product_Category', hue='Product_Category', data=df, palette='Set2', legend=False)

sns.set(rc={'figure.figsize':(40,5)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(35,5)})
sns.barplot(x='Product_Category', y='Amount', data=sales_product_category, hue='Product_Category')
plt.show()
```

```python
sales_product_category = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_ID', y='Orders', data=sales_product_category, hue='Product_ID')
plt.show()
```
